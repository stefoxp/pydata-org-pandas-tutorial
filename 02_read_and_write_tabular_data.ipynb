# How do I read and write tabular data?

Data used for this tutorial:

titanic.csv

```python
import pandas as pd
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
titanic = pd.read_csv(io.StringIO(uploaded['titanic.csv'].decode('utf-8')))
```

```python
titanic
```

```python
# see the first 8 row
titanic.head(8)
```

```python
# columns data types
titanic.dtypes
```

```python
# transform to spreadsheet
titanic.to_excel("titanic.xlsx", sheet_name="passengers", index=False)
```

```python
# reload the data from spreadsheet
titanic_ss = pd.read_excel("titanic.xlsx", sheet_name="passengers")
# data
titanic_ss
```
